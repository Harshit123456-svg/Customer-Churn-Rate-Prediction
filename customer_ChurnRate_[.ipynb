```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.tree import DecisionTreeClassifier

# Calculate Mutual Information scores for each feature

# Separate features (X) and target (y)
X = df_N  # Numerical columns in df_N
y = df['Churn']               # Get the 'target' column as the label

# Calculate Mutual Information scores for each feature
mutual_info = SelectKBest(mutual_info_classif, k='all')
mutual_info.fit(X, y)

# Get the scores and corresponding feature names
feature_scores = mutual_info.scores_
feature_names = X.columns

# Create a DataFrame to store the scores and feature names
mutual_info_df = pd.DataFrame({'Feature': feature_names, 'Mutual_Info_Score': feature_scores})

# Sort the DataFrame by Mutual Information scores in descending order
mutual_info_df = mutual_info_df.sort_values(by='Mutual_Info_Score', ascending=False)

# Display the top 5 features with the highest Mutual Information scores
top_5_mutual_info = mutual_info_df.head(5)
print("Top 5 Features by Mutual Information (Information Gain):")
print(top_5_mutual_info)

# Calculating Gini Index
# Initialize a Decision Tree classifier
clf = DecisionTreeClassifier(random_state=0)

# Fit the classifier to your data
clf.fit(X, y)

# Calculate Gini Importance scores for each feature
gini_importance = clf.feature_importances_

# Create a DataFrame to store the Gini Importance scores and feature names
gini_df = pd.DataFrame({'Feature': feature_names, 'Gini_Importance': gini_importance})

# Sort the DataFrame by Gini Importance scores in descending order
gini_df = gini_df.sort_values(by='Gini_Importance', ascending=False)

# Display the top 5 features with the highest Gini Importance scores
top_5_gini = gini_df.head(5)
print("\nTop 5 Features by Gini Index:")
print(top_5_gini)
```

```python
import pandas as pd
import nltk
nltk.download('punkt')
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.util import ngrams

# Function to perform stemming for a text column
def perform_stemming(text_column):
    stemmer = PorterStemmer()
    return text_column.apply(lambda x: ' '.join([stemmer.stem(word) for word in nltk.word_tokenize(x)]))

# Function to perform lemmatization for a text column
def perform_lemmatization(text_column):
    lemmatizer = WordNetLemmatizer()
    return text_column.apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in nltk.word_tokenize(x)]))

# Function to generate n-grams for a text column
def generate_ngrams(text_column, n):
    return text_column.apply(lambda x: [' '.join(gram) for gram in ngrams(nltk.word_tokenize(x), n)])

# Define the text columns in your DataFrame
text_columns = df.select_dtypes(include=['object'])

# Perform stemming for all text columns
for col in text_columns.columns:
    df[col + '_Stemmed'] = perform_stemming(df[col])

# Perform lemmatization for all text columns
for col in text_columns.columns:
    df[col + '_Lemmatized'] = perform_lemmatization(df[col])

# Generate bigrams for all text columns
n = 2  # Change n to generate different n-grams (e.g., n=2 for bigrams, n=3 for trigrams)
for col in text_columns.columns:
    df[col + '_Bigrams'] = generate_ngrams(df[col], n)

# Display the updated DataFrame with stemmed, lemmatized, and n-gram columns
df.head(5)
```
